```python
import pandas as pd

```

Read The File

```python
df=pd.read_excel("Applicants.xlsx")
df.head(5)
```

Check the data type for each Column

```python
df.info()
```

Check num of Null Value for Each Column

```python
df.isnull().sum()
```

 Replace hyphens with an empty string

```python
df['Contact number'] = df['Contact number'].str.replace('-', '')
```

Reorder The names

```python
df['first_part'] = df['Modified By'].str.split(',').str[0]  # Part before the comma
df['last_part'] = df['Modified By'].str.split(',').str[1].str.strip()  # Part after the comma

# Rejoin the parts in "First Last" format
df['Modified By'] = df['last_part'] + ' ' + df['first_part']  # Rejoin in desired format
```

```python
df['first_part1'] = df['Created By'].str.split(',').str[0]  # Part before the comma
df['last_part2'] = df['Created By'].str.split(',').str[1].str.strip()  # Part after the comma

# Rejoin the parts in "First Last" format
df['Created By'] = df['last_part2'] + ' ' + df['first_part1']  # Rejoin in desired format
```

```python
df.drop(columns=['last_part', 'first_part', 'first_part1', 'last_part2'], inplace=True)
```

--------------------------------------------------------------------------------

Convert Data type from Object To DateTime


```python
df['Technical interview date'] = pd.to_datetime(df['Technical interview date'])
df['Board interview date'] = pd.to_datetime(df['Board interview date'])
df['Modified'] = pd.to_datetime(df['Modified'])
df['Created'] = pd.to_datetime(df['Created'])
```

```python
df['Modified'] = df['Modified'].dt.strftime('%m/%d/%Y  %H:%M')  # 24-hour format
df['Technical interview date'] =  df['Technical interview date'].dt.strftime('%m/%d/%Y  %H:%M')
df['Board interview date'] = df['Board interview date'].dt.strftime('%m/%d/%Y  %H:%M')
df['Created'] = df['Created'].dt.strftime('%m/%d/%Y  %H:%M')

```

```python
df.info()
```

--------------------------------------------------------------------------------

Fill Numeric and Str Null Value

```python
df['fit_to_account'] = df['fit_to_account'].replace("[]", "No fitting to any account")
df['Applicant Name'] = df['Applicant Name'].fillna("Null Applicant Name Value")
df['Role ID'] = df['Role ID'].fillna(0000)  # Fill NaN with the integer value 000000
df['Level ID'] = df['Level ID'].fillna(0000)
df['Years of experience'] = df['Years of experience'].fillna(0000)
df['Contact number'] = df['Contact number'].fillna(00000000000)
df['Interviewers names'] = df['Interviewers names'].fillna("Null Interviewers names Value")

```

```python
df.head(5)
```

Fill Datetime Null Values With default Value "2000-01-01 00:00:00"

```python
df['Technical interview date'] = df['Technical interview date'].fillna("01/01/2000 00:00")
df['Board interview date'] = df['Board interview date'].fillna("01/01/2000 00:00")
```

```python
df.head(5)
```

```python
df.info()
```

```python
file_path = r"Applicants1.xlsx"


try:
    df.to_excel(file_path, index=False)
    print("File saved successfully.")
except Exception as e:
    print("Error saving file:", e)
```
